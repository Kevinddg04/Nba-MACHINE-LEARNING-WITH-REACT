```python
# ---------------------------
# BLOQUE 1: Ingeniería de features (DF -> DF_points_clean)
# - merge del rival por gameId
# - shift + rolling (attack r5, defense r10)
# - luego dropear las columnas crudas (raw) que ya no se necesitan
# ---------------------------

# Partimos del df original (con columnas crudas)
dfp = df.copy()

# Asegurar gameId existe (si no tienes, crea como mostramos antes)
if "gameId" not in dfp.columns:
    dfp["gameId"] = (
        dfp["gameDateTimeEst"].astype(str) + "_" +
        dfp[["teamId","opponentTeamId"]].min(axis=1).astype(str) + "_" +
        dfp[["teamId","opponentTeamId"]].max(axis=1).astype(str)
    )

# 1) Orden
dfp = dfp.sort_values(["teamId", "gameDateTimeEst"]).reset_index(drop=True)

# 2) Crear tabla rival (las columnas crudas del partido que tiene el rival)
rival_cols = [
    'teamId', 'gameId', 'teamScore', 'assists', 'blocks', 'steals',
    'fieldGoalsAttempted', 'fieldGoalsMade', 'fieldGoalsPercentage',
    'threePointersAttempted', 'threePointersMade', 'threePointersPercentage',
    'freeThrowsAttempted', 'freeThrowsMade', 'freeThrowsPercentage',
    'reboundsDefensive', 'reboundsOffensive', 'reboundsTotal',
    'foulsPersonal', 'turnovers'
]

# Asegurarse de que están en dfp antes de seleccionar
rival_cols_exist = [c for c in rival_cols if c in dfp.columns]
rival = dfp[rival_cols_exist].rename(columns=lambda c: c + "_OPP")

# 3) Merge rival con dfp para obtener stats del oponente en la misma fila
#    left_on = (gameId, opponentTeamId) ; right_on = (gameId_OPP, teamId_OPP)
dfp = dfp.merge(
    rival,
    left_on=["gameId", "opponentTeamId"],
    right_on=["gameId_OPP", "teamId_OPP"],
    how="left"
)

# 4) Definir columnas para rolling (ataque = columnas del equipo; defensa = columnas del oponente)
attack_cols = [
    'teamScore', 'assists', 'blocks', 'steals',
    'fieldGoalsAttempted', 'fieldGoalsMade',
    'threePointersAttempted', 'threePointersMade',
    'freeThrowsAttempted', 'freeThrowsMade',
    'reboundsOffensive', 'turnovers'
]
# dejar solo las que existan
attack_cols = [c for c in attack_cols if c in dfp.columns]

defense_cols = [
    'teamScore_OPP', 'assists_OPP', 'blocks_OPP', 'steals_OPP',
    'fieldGoalsAttempted_OPP', 'fieldGoalsMade_OPP',
    'threePointersAttempted_OPP', 'threePointersMade_OPP',
    'freeThrowsAttempted_OPP', 'freeThrowsMade_OPP',
    'reboundsOffensive_OPP', 'turnovers_OPP'
]
defense_cols = [c for c in defense_cols if c in dfp.columns]

# 5) Rolling ofensivo (r5) — shift antes de rolling para evitar leakage
for col in attack_cols:
    dfp[f"ATT_{col}_r5"] = (
        dfp.groupby("teamId")[col]
        .shift(1)                      # no usar el partido actual
        .rolling(5, min_periods=1)
        .mean()
        .reset_index(level=0, drop=True)
    )

# 6) Rolling defensivo (r10) — shift antes de rolling para evitar leakage
for col in defense_cols:
    dfp[f"DEF_{col}_r10"] = (
        dfp.groupby("teamId")[col]
        .shift(1)
        .rolling(10, min_periods=1)
        .mean()
        .reset_index(level=0, drop=True)
    )

# 7) Ahora que ya tienes los rollings, dropear las columnas CRUDAS que generan leakage.
#    Este es el drop de ingeniería de features (lo que tú llamaste drop_cols).
#    NOTA: se borran las columnas crudas del partido actual y las columnas crudas del rival (_OPP)
drop_cols = [
    'opponentScore',
    # columnas crudas del rival (si existen)
    'teamScore_OPP',
    'assists_OPP', 'blocks_OPP', 'steals_OPP',
    'fieldGoalsAttempted_OPP', 'fieldGoalsMade_OPP', 'fieldGoalsPercentage_OPP',
    'threePointersAttempted_OPP', 'threePointersMade_OPP', 'threePointersPercentage_OPP',
    'freeThrowsAttempted_OPP', 'freeThrowsMade_OPP', 'freeThrowsPercentage_OPP',
    'reboundsOffensive_OPP', 'turnovers_OPP'
]
# eliminar solo las que existan
drop_cols = [c for c in drop_cols if c in dfp.columns]
dfp = dfp.drop(columns=drop_cols, errors="ignore")

# 8) Opcional: quitar columnas de porcentaje crudas del partido actual si aún existen (leakage)
for c in ['fieldGoalsPercentage','threePointersPercentage','freeThrowsPercentage']:
    if c in dfp.columns:
        dfp = dfp.drop(columns=[c], errors="ignore")

# 9) Validación: asegurarse de que aparecen las columnas ATT_ y DEF_ calculadas
att_cols_calc = [c for c in dfp.columns if c.startswith("ATT_")]
def_cols_calc = [c for c in dfp.columns if c.startswith("DEF_")]

if len(att_cols_calc) == 0:
    raise RuntimeError("No se crearon columnas ATT_. Revisa attack_cols y el merge.")
if len(def_cols_calc) == 0:
    # Esto puede ocurrir si no había columnas OPP en el df para calcular defensa;
    # en ese caso la defensa tendrá que construirse de otra forma.
    print("Warning: no se detectaron columnas DEF_. Verifica que el merge aportó columnas _OPP.")

print("Bloque 1 completado. ATT cols:", len(att_cols_calc), "DEF cols:", len(def_cols_calc))
dfp.dropna(inplace=True)
dfp.shape

```

```python
# ===========================================================
#  PREPARACIÓN DE DATOS SIN CREAR COLUMNAS NUEVAS
#  Se reemplazan teamScore, opponentScore, PointDiff, Handicap,
#  y se crean solo las columnas estrictamente necesarias
# ===========================================================

df = df.sort_values(["teamId", "gameDateTimeEst"]).reset_index(drop=True)
df["totalPoints_real"] = df["teamScore"] + df["opponentScore"]

# ===========================================================
# 1️⃣ SHIFT – estadísticas del partido anterior (EVITA LEAKING)
# ===========================================================
shift_cols = [
    "teamScore", "opponentScore",
    "assists", "blocks", "steals",
    "fieldGoalsAttempted", "fieldGoalsMade", "fieldGoalsPercentage",
    "threePointersAttempted", "threePointersMade", "threePointersPercentage",
    "freeThrowsAttempted", "freeThrowsMade", "freeThrowsPercentage",
    "reboundsDefensive", "reboundsOffensive", "reboundsTotal",
    "foulsPersonal", "turnovers",
]

df[shift_cols] = df.groupby("teamId")[shift_cols].shift(1)

# ===========================================================
# 2️⃣ ROLLING — promedio últimos 5 partidos (OPTIMIZADO)
#    → Se reescriben las mismas columnas, NO se crean nuevas
# ===========================================================
rolling_cols = [
    "teamScore", "opponentScore",
    "assists", "blocks", "steals",
    "fieldGoalsAttempted", "fieldGoalsMade",
    "threePointersAttempted", "threePointersMade",
    "freeThrowsAttempted", "freeThrowsMade",
    "reboundsDefensive", "reboundsOffensive", "reboundsTotal",
    "foulsPersonal", "turnovers",
]

df[rolling_cols] = (
    df.groupby("teamId")[rolling_cols]
      .rolling(5, min_periods=1)
      .mean()
      .reset_index(level=0, drop=True)
)

# ===========================================================
# 3️⃣ CALCULAR DIFERENCIA DE PUNTOS (HISTÓRICA, SIN LEAKING)
# ===========================================================
df["PointDiff"] = df["teamScore"] - df["opponentScore"]
# ===========================================================
# 4️⃣ CALCULAR HANDICAP HISTÓRICO
#     → Handicap = lo que debe cubrir el equipo
# ===========================================================
df["Handicap"] = df["PointDiff"] * -1

# ===========================================================
# 5️⃣ EXPECTED SCORES (rolling 5 partidos) — OPTIMIZADOS
# ===========================================================
df["expectedTeamScore"] = df.groupby("teamId")["teamScore"].transform(
    lambda x: x.rolling(5, min_periods=1).mean()
)

df["expectedOpponentScore"] = df.groupby("teamId")["opponentScore"].transform(
    lambda x: x.rolling(5, min_periods=1).mean()
)
#Creacion totalPoints
df["totalPoints"] = df["expectedTeamScore"] + df["expectedOpponentScore"]

# ===========================================================
# 6️⃣ STREAKS (rachas de victorias) — CORREGIDO (SIN LEAKING)
# ===========================================================
df["win"] = df["win"].astype(int)

df["win_streak_5"] = (
    df.groupby("teamId")["win"]
      .apply(lambda x: x.shift(1).rolling(5, min_periods=1).sum())
      .reset_index(level=0, drop=True)
)

# ===========================================================
# 7️⃣ CALCULAR REAL HANDICAP (VERSIÓN REAL DEL PARTIDO)
#     IMPORTANTE: shift para evitar leakage
# ===========================================================
df["RealHandicap"] = df["teamScore"] - df["opponentScore"]
df["RealHandicap"] = (
    df.groupby("teamId")["RealHandicap"]
      .apply(lambda s: s.shift(1).rolling(5, min_periods=1).mean())
      .reset_index(level=0, drop=True)
)


# ===========================================================
# 8️⃣ LIMPIEZA FINAL (primeros partidos con NaN)
# ===========================================================
df = df.dropna().reset_index(drop=True)



```

```python
# =======================================
# 1️⃣ Librerías
# =======================================
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# =======================================
# 3️⃣ Variable objetivo
# =======================================
y = df['win']


# =======================================
# 4️⃣ Selección final de features
# =======================================
features = [

    "assists", "blocks", "steals",
    "fieldGoalsAttempted", "fieldGoalsMade", "fieldGoalsPercentage",
    "threePointersAttempted", "threePointersMade", "threePointersPercentage",
    "freeThrowsAttempted", "freeThrowsMade", "freeThrowsPercentage",
    "reboundsDefensive", "reboundsOffensive", "reboundsTotal",
    "foulsPersonal", "turnovers",'totalPoints','RealHandicap',

    # Rolling correcto que sí existe
    "win_streak_5",

    # Tus features calculados
    "expectedTeamScore",
    "expectedOpponentScore",
]

# Validación automática para evitar errores
features = [f for f in features if f in df.columns]

X = df[features]


# =======================================
# 5️⃣ Train/Test Split
# =======================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)


# =======================================
# 6️⃣ Entrenar CatBoost
# =======================================
model = CatBoostClassifier(
    iterations=1200,
    learning_rate=0.045,
    depth=6,
    eval_metric='Accuracy',
    random_seed=42,
    verbose=100,
    use_best_model=True
)

model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test)
)


# =======================================
# 7️⃣ Evaluación
# =======================================
y_pred = model.predict(X_test)

print("\nAccuracy:", round(accuracy_score(y_test, y_pred) * 100, 2),"%")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))


# =======================================
# 8️⃣ Importancia de Features
# =======================================
importances = model.get_feature_importance(prettified=True)
print("\nIMPORTANCIA DE VARIABLES:\n", importances)
```

```python
# ---------------------------
# BLOQUE 2: creación del target y entrenamiento CatBoost
# - target: teamScore_next (predicción del próximo partido del equipo)
# - usamos DF_points_clean (DFP con ATT_/DEF_ calculadas)
# ---------------------------

from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# =============================================================
# 1) Crear target 'teamScore_next' desde DFP (dataset de features)
#    NOTA: NO MODIFICAMOS dfp. Solo lo leemos.
# =============================================================
df_next = dfp.sort_values(["teamId", "gameDateTimeEst"]).copy()
df_next["teamScore_next"] = dfp.groupby("teamId")["teamScore"].shift(-1)

# quedarnos solo con columnas necesarias
df_next = df_next[["gameId", "teamId", "teamScore_next"]]

# =============================================================
# 2) Crear DF_MODEL (NO reescribimos dfp para evitar confusión)
# =============================================================
df_model = dfp.merge(df_next, on=["gameId", "teamId"], how="left")

# eliminar filas sin target
df_model = df_model.dropna(subset=["teamScore_next"]).reset_index(drop=True)

# =============================================================
# 3) Eliminar metadata innecesaria
# =============================================================
cols_to_remove = [
    'index', 'gameDateTimeEst',
    'teamId_OPP', 'gameId_OPP',
    'teamId', 'opponentTeamId', 'gameId',
    'home', 'win'
]

df_model = df_model.drop(
    columns=[c for c in cols_to_remove if c in df_model.columns],
    errors='ignore'
)

# =============================================================
# 4) Seleccionar features (ATT_ y DEF_)
# =============================================================
feature_cols = [c for c in df_model.columns if c.startswith("ATT_") or c.startswith("DEF_")]

X = df_model[feature_cols]
y = df_model["teamScore_next"]

# =============================================================
# 5) Split temporal (80/20)
# =============================================================
split = int(len(df_model) * 0.80)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# =============================================================
# 6) Entrenar CatBoost
# =============================================================
model = CatBoostRegressor(
    iterations=1200,
    depth=6,
    learning_rate=0.03,
    loss_function="RMSE",
    verbose=200,
    random_seed=42
)

model.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=200)

# =============================================================
# 7) Evaluación
# =============================================================
preds = model.predict(X_test)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print(f"MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# =============================================================
# 8) Guardar referencias
# =============================================================
FEATURE_COLS_USED = feature_cols
DF_MODEL = df_model.copy()
MODEL_TEAMPOINTS = model

# =============================================================
# 9) Importancia de features
# =============================================================
importances = model.get_feature_importance()
feat_imp_df = pd.DataFrame({
    "Feature": feature_cols,
    "Importance": importances
}).sort_values("Importance", ascending=False)

print("\nIMPORTANCIA DE FEATURES:")
print(feat_imp_df)

print("Bloque 2 completado. Modelo entrenado.")







```
